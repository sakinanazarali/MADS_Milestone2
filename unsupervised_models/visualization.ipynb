```python
df_pca_k_means = pd.DataFrame(pca_k_means_data)
df_tsne_k_means = pd.DataFrame(tsne_k_means_data)
df_pca_agglomerative = pd.DataFrame(pca_agglomerative_data)
df_tsne_agglomerative = pd.DataFrame(tsne_agglomerative_data)
```

```python
df_combined_k_means = pd.concat([df_pca_k_means, df_tsne_k_means])

avg_metrics = df_combined_k_means.groupby('Method').mean().reset_index()

plt.figure(figsize=(14, 8))

plt.subplot(2, 2, 1)
sns.barplot(x='Method', y='Silhouette Score', data=avg_metrics)
plt.title('Average Silhouette Score')

plt.subplot(2, 2, 2)
sns.barplot(x='Method', y='Davies-Bouldin Index', data=avg_metrics)
plt.title('Average Davies-Bouldin Index')

plt.subplot(2, 2, 3)
sns.barplot(x='Method', y='Calinski-Harabasz Index', data=avg_metrics)
plt.title('Average Calinski-Harabasz Index')

plt.subplot(2, 2, 4)
sns.barplot(x='Method', y='Percentage of Similarity', data=avg_metrics)
plt.title('Average Percentage of Similarity')

plt.tight_layout()
plt.show()
```

```python
df_combined_agg = pd.concat([df_tsne_agglomerative, df_pca_agglomerative])

avg_metrics = df_combined_agg.groupby('Method').mean().reset_index()


plt.figure(figsize=(14, 8))

plt.subplot(2, 2, 1)
sns.barplot(x='Method', y='Silhouette Score', data=avg_metrics)
plt.title('Average Silhouette Score')

plt.subplot(2, 2, 2)
sns.barplot(x='Method', y='Davies-Bouldin Index', data=avg_metrics)
plt.title('Average Davies-Bouldin Index')

plt.subplot(2, 2, 3)
sns.barplot(x='Method', y='Calinski-Harabasz Index', data=avg_metrics)
plt.title('Average Calinski-Harabasz Index')

plt.subplot(2, 2, 4)
sns.barplot(x='Method', y='Percentage of Similarity', data=avg_metrics)
plt.title('Average Percentage of Similarity')

plt.tight_layout()
plt.show()
```
